## To find x - x_bar

```python
# Manual calculation
```

```python
x
```

```python
mean
```

```python
x1 = []
```

```python
for i in x:
    x1.append(i-mean)
```

```python
x1
```

```python
# To find x-x_bar the whole square
```

```python
square = 0
```

```python
for i in x1:
    square += i*i
```

```python
square
```

```python
square = 4.0 + 1.0 + 0.0 + 1.0 + 4.0
```

```python
var
```
