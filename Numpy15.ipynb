# Covariance

# If we are having Economic growth data with the rate of interest, ecg and roi. See how will calculate using Covariance 

```python
import numpy as np
```

```python
ecg=[2.1,2.5,4.0,3.6]
```

```python
roi=[8,12,14,10]
```

```python
mean_ecg=np.mean(ecg)
```

```python
mean_ecg
```

```python
mean_roi=np.mean(roi)
```

```python
mean_roi
```

```python
ecg1 = []
```

```python
for i in ecg:
    ecg1.append(i - mean_ecg)
```

```python
ecg1
```

```python
roi1 = []
```

```python
for i in roi:
    roi1.append(i - mean_roi)
```

```python
roi1
```

```python
# Now we find out x-x_bar and y-y_bar
```

```python
# Now multiply x-x_bar * y-y_bar
```

```python
-0.9499999999999997 * -3.0
```

```python
 -0.5499999999999998 * 1.0
```

```python
 0.9500000000000002 * 3.0
```

```python
 0.5500000000000003 * -1.0
```

```python
res = np.multiply(ecg1,roi1) # cov(x-x_bar,y-y_bar)
```

```python
res
```

```python
N = len(ecg)  # Number of data points
```

```python
N
```

```python
# Calculate the covariance with Bessel correction
```

```python
cov = np.sum(res)/(N - 1)
```

```python
cov
```

```python
np.cov(ecg,roi)
```

```python
np.cov(ecg, roi)[0, 1]
```
