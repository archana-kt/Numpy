# Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random samples from a normal distribution with mean 10 and standard deviation 2
x = np.random.normal(10, 2, size=1000)
```

```python
x
```

```python
# Calculate mean and standard deviation
mean = np.mean(x)
std_dev = np.std(x)
var = np.var(x)
```

```python
# Create an array of values for the x-axis
x_values = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
```

```python
# Calculate the corresponding y-values using the normal distribution formula
y_values = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - mean)/std_dev)**2)
```

y_values:

y_values is an array of values representing the y-coordinates (vertical axis) for the normal distribution plot.
These values are calculated using the formula for the probability density function (PDF) of a normal distribution:
    

```python
# Plotting the bell curve
plt.plot(x_values, y_values)
plt.show()
```
